### 问题1： 输出所有由四个数字1、2、3、5组出的数字组

```python
#  答案:
def combinations(list_1):
    for i in list_1:
        for j in list_1:
            for k in list_1:
                for l in list_1:
                    print(i,j,k,l)
combinations([1,2,3,5])
```

### 问题2： 输出斐波那契数列前20项的和

```python
#  答案:
function_2 = []
for i in range(0,20):
    if i == 0 or i == 1:
        function_2.append(1)
    else:
        function_2.append(function_2[i-2] + function_2[i-1])
print(function_2)
```

### 问题3： 输出200以内，各个位数之和为偶数的素数（如101、103）

```python
#  答案:
for i in range(2,200):
    k=True
    for j in range(2,i):
        if(i%j==0):
            k=False
            break
    if(k==True):
        hund=int(i/100)
        ten=int((i-hund*100)/10)
        one=i-hund*100-ten*10
        sum=hund+ten+one
        if sum%2==0:
            print (hund*100+ten*10+one)
```
